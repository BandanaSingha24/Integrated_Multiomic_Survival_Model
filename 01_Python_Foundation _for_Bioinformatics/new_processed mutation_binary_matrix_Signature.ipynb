```python
import pandas as pd

# Step 1: Load the raw mutation data (कच्चे म्यूटेशन डेटा को लोड करें)
raw_mut_df = pd.read_csv('/content/data_mutations.txt', sep='\t', skiprows=1)

# Step 2: Filter out 'Silent' mutations to reduce noise (शोर कम करने के लिए 'Silent' म्यूटेशन को हटा दें)
mut_cleaned = raw_mut_df[~raw_mut_df['Variant_Classification'].isin(['Silent'])]

# Step 3: Create a Binary Matrix using Patient IDs and Gene Symbols (पेशेंट आईडी और जीन सिंबल का उपयोग करके बाइनरी मैट्रिक्स बनाएं)
# This aligns mutations to specific patients (यह म्यूटेशन को विशिष्ट मरीजों से जोड़ता है)
final_matrix = pd.crosstab(mut_cleaned['Tumor_Sample_Barcode'], mut_cleaned['Hugo_Symbol'])

# Step 4: Ensure the matrix contains only 0 and 1 (सुनिश्चित करें कि मैट्रिक्स में केवल 0 और 1 ही हों)
# 1 indicates presence of mutation, 0 indicates absence (1 म्यूटेशन की उपस्थिति को दर्शाता है, 0 अनुपस्थिति को)
final_binary_matrix = (final_matrix > 0).astype(int)

# Step 5: Check the shape to verify data integrity (डेटा की अखंडता की जांच करने के लिए शेप देखें)
print("Binary Matrix Shape:", final_binary_matrix.shape)
print(final_binary_matrix.head())
```

```python
# Save the binary mutation matrix as a CSV file (बाइनरी म्यूटेशन मैट्रिक्स को CSV फाइल के रूप में सेव करें)
final_binary_matrix.to_csv('processed_mutation_ready_for_integration.csv')

# Use this to download the file to your computer (फाइल को अपने कंप्यूटर में डाउनलोड करने के लिए इसका उपयोग करें)
from google.colab import files
files.download('processed_mutation_ready_for_integration.csv')
```

```python
# Step 1: Data Preparation and Raw Data Extraction
# (स्टेप 1: डेटा की तैयारी और रॉ डेटा एक्सट्रैक्शन)

# 1.1 Load the raw mutation file
# (1.1 रॉ म्यूटेशन फाइल को लोड करें)
raw_mut_df = pd.read_csv('/content/data_mutations.txt', sep='\t', skiprows=1)

# 1.2 Save the raw data for future Mutational Signature analysis (Module 4)
# (1.2 भविष्य के म्यूटेशनल सिग्नेचर विश्लेषण (मॉड्यूल 4) के लिए रॉ डेटा को सेव करें)
raw_mut_df.to_csv('raw_mutation_for_signature.csv', index=False)

# 1.3 Filter out 'Silent' mutations to ensure high data quality
# (1.3 डेटा की उच्च गुणवत्ता सुनिश्चित करने के लिए 'Silent' म्यूटेशन को फिल्टर करें)
mut_cleaned = raw_mut_df[~raw_mut_df['Variant_Classification'].isin(['Silent'])]

# 1.4 Create a Binary Matrix for integration and survival analysis (Module 5 & 6)
# (1.4 इंटीग्रेशन और सर्वाइवल एनालिसिस (मॉड्यूल 5 और 6) के लिए एक बाइनरी मैट्रिक्स बनाएं)
binary_matrix = pd.crosstab(mut_cleaned['Tumor_Sample_Barcode'], mut_cleaned['Hugo_Symbol'])
binary_matrix = (binary_matrix > 0).astype(int)

# 1.5 Save the processed binary matrix for later use
# (1.5 बाद में उपयोग के लिए प्रोसेस्ड बाइनरी मैट्रिक्स को सेव करें)
binary_matrix.to_csv('binary_mutation_for_integration.csv')

print("Process Completed: Both files (Raw and Binary) have been saved successfully.")
print("(प्रक्रिया पूरी हुई: दोनों फाइलें (रॉ और बाइनरी) सफलतापूर्वक सेव हो गई हैं।)")
```
