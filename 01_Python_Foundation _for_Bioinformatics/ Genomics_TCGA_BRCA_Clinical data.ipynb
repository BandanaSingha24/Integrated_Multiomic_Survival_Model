```python
import pandas as pd
import numpy as np

 #step 1 : lodind the Clinical data

# 1.1 Load the patient clinical data
# comment='#' skips the metadata/header rows starting with a hash symbol
# comment='#' का उपयोग करके हम उन लाइनों को छोड़ रहे हैं जो '#' से शुरू होती हैं
patient_df = pd.read_csv('data_clinical_patient.txt', sep='\t', comment='#')

# 1.2 Load the sample clinical data
sample_df = pd.read_csv('data_clinical_sample.txt', sep='\t', comment='#')

#  Display the first 5 rows of both DataFrames to verify the data
print("--- Patient Data Overview ---")
display(patient_df.head())

print("\n--- Sample Data Overview ---")
display(sample_df.head())
```

```python
import pandas as pd

# Step 1: Load Gene Expression Data

# Load the mRNA expression data using tab separator
mrna_df = pd.read_csv('data_mrna_illumina_microarray_zscores_ref_diploid_samples.txt', sep='\t')

# Display the first 5 rows to verify the data structure
print("--- Gene Expression Data Overview ---")
display(mrna_df.head())
```

```python
import pandas as pd

# Step 1: Load Mutation Data
# Load the mutation data using tab separator, skipping comment lines starting with '#'
mutation_df = pd.read_csv('data_mutations.txt', sep='\t', comment='#')

# Display the first 5 rows to verify the data structure
print("--- Mutation Data Overview ---")
display(mutation_df.head())
```

```python
# step 1: Loading mutation data properly by skipping header metadata
# म्यूटेशन डेटा को सही तरीके से लोड करें और ऊपर की एक्स्ट्रा लाइन्स को हटा दें
mutation_df = pd.read_csv('data_mutations.txt', sep='\t', skiprows=1)

# Check the column names to ensure 'Tumor_Sample_Barcode' exists
print("Available columns:", mutation_df.columns.tolist())
```

```python
# #Step 2 : Merging the clinical datasets
# हिंदी: #Step 2 : क्लीनिकल डेटासेट्स को आपस में मर्ज करना

# 2.1 Merge patient and sample dataframes using 'PATIENT_ID' as the common key
combined_df = pd.merge(patient_df, sample_df, on='PATIENT_ID', how='inner')

# 2.2 Print the shape of the merged dataframe to see total rows and columns
print("--- Merged Data Shape ---")
print(f"Total Rows (Patients/Samples): {combined_df.shape[0]}")
print(f"Total Columns (Features): {combined_df.shape[1]}")

# 2.3 Display the first 5 rows of the combined dataframe to verify the merge
print("\n--- Combined Data Overview ---")
display(combined_df.head())
```

```python
# Step 2: Transpose Gene Expression Data
# (Transpose to bring genes to columns)
mrna_df = mrna_df.set_index('Hugo_Symbol').T

# Set Patient ID as index for alignment
mrna_df.index.name = 'PATIENT_ID'
mrna_df.reset_index(inplace=True)

print("--- Transposed Gene Expression Data Overview ---")
display(mrna_df.head())
```

```python
# step 2 Mutation matrix

# Removing unwanted space from column
mutation_df.columns = mutation_df.columns.str.strip()


# अब बाइनरी मैट्रिक्स बनाएं
mutation_matrix = pd.crosstab(mutation_df['Tumor_Sample_Barcode'], mutation_df['Hugo_Symbol'])
mutation_matrix = (mutation_matrix > 0).astype(int)

print("--- Binary Mutation Matrix Overview ---")
display(mutation_matrix.head())
```

```python
# #Step 3 : Checking for Missing Values

# 3.1 Calculate total missing values for each column
missing_info = combined_df.isnull().sum()

# 3.2 Filter and display only columns that have at least one missing value
# हिंदी: 3.2 केवल उन कॉलम्स को फिल्टर करके देखना जिनमें कम से कम एक मिसिंग वैल्यू है
missing_only = missing_info[missing_info > 0]

print("--- Columns with Missing Values ---")
print(missing_only)
```

```python
# #Step 4 : Handling Missing Values (Imputation) - Update
# (not important we can apply automatic imputation directly)
# 4.1 Handle Categorical columns by replacing NaN with 'Unknown'
categorical_cols = [
    'CELLULARITY', 'CHEMOTHERAPY', 'ER_IHC', 'HER2_SNP6',
    'HORMONE_THERAPY', 'INFERRED_MENOPAUSAL_STATE', 'INTCLUST',
    'OS_STATUS', 'CLAUDIN_SUBTYPE', 'THREEGENE', 'VITAL_STATUS',
    'LATERALITY', 'RADIO_THERAPY', 'HISTOLOGICAL_SUBTYPE', 'BREAST_SURGERY',
    'RFS_STATUS', 'ER_STATUS', 'HER2_STATUS', 'PR_STATUS'
]
combined_df[categorical_cols] = combined_df[categorical_cols].fillna('Unknown')

# 4.2 Handle Numerical columns by replacing NaN with the Median value
numerical_cols = [
    'LYMPH_NODES_EXAMINED_POSITIVE', 'NPI', 'COHORT',
    'AGE_AT_DIAGNOSIS', 'OS_MONTHS', 'RFS_MONTHS',
    'TUMOR_SIZE', 'TUMOR_STAGE'
]
for col in numerical_cols:
    median_value = combined_df[col].median()
    combined_df[col] = combined_df[col].fillna(median_value)

# 4.3 Re-check if any missing values are left in the entire dataset
remaining_missing = combined_df.isnull().sum().sum()
print("--- Missing Values Verification ---")
print(f"Total remaining missing values in the entire dataframe: {remaining_missing}")
```

```python
# #Step 4 : Final Automatic Imputation to Fix Remaining 121 Missing Values

# 4.1 Automatically identify categorical and numerical columns
cat_cols = combined_df.select_dtypes(include=['object']).columns
num_cols = combined_df.select_dtypes(include=['int64', 'float64']).columns

# 4.2 Fill Categorical missing values with 'Unknown'
combined_df[cat_cols] = combined_df[cat_cols].fillna('Unknown')

# 4.3 Fill Numerical missing values with Median after forcing numeric conversion
for col in num_cols:
    combined_df[col] = pd.to_numeric(combined_df[col], errors='coerce')
    median_val = combined_df[col].median()
    combined_df[col] = combined_df[col].fillna(median_val)

# 4.4 Final verification check
final_missing = combined_df.isnull().sum().sum()
print("--- Final Missing Values Verification ---")
print(f"Total remaining missing values now: {final_missing}")
```

```python
# #Step 5 : Encoding Categorical Receptors to Numbers (0 and 1)

# 5.1 Define a mapping dictionary for Positive and Negative status
# हिंदी: 5.1 Positive और Negative स्टेटस के लिए एक मैपिंग (नंबर) तय करना
status_mapping = {'Positive': 1, 'Negative': 0, 'Unknown': 0}

# 5.2 Apply mapping to the main breast cancer receptor columns
receptor_cols = ['ER_STATUS', 'PR_STATUS', 'HER2_STATUS']

for col in receptor_cols:
    combined_df[col] = combined_df[col].map(status_mapping)

# 5.3 Verify the changes by looking at these specific columns
print("--- Encoded Receptors Overview ---")
display(combined_df[['PATIENT_ID'] + receptor_cols].head())
```

```python
from sklearn.model_selection import train_test_split

# #Step 6 : Final Data Preparation for Machine Learning (supervised)

# 6.1 Select the clinical features we want to use for our survival prediction model
# हिंदी: 6.1 उन क्लीनिकल फीचर्स को चुनना जिनका उपयोग हम मॉडल के लिए करना चाहते हैं
features_to_keep = [
    'AGE_AT_DIAGNOSIS', 'TUMOR_SIZE', 'TUMOR_STAGE',
    'ER_STATUS', 'PR_STATUS', 'HER2_STATUS', 'CELLULARITY', 'CHEMOTHERAPY'
]

# 6.2 Define X (Features) and y (Target variable - Survival Status)
X = combined_df[features_to_keep]
y = combined_df['OS_STATUS'].map({'Living': 1, 'Deceased': 0}) # 1 = Living, 0 = Died

# 6.3 Apply One-Hot Encoding to categorical columns like CELLULARITY and CHEMOTHERAPY
X_encoded = pd.get_dummies(X, columns=['CELLULARITY', 'CHEMOTHERAPY'], drop_first=True)

# 6.4 Split the dataset into 80% Training and 20% Testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# 6.5 Print the final shapes of our datasets
print("--- Data Splitting Completed ---")
print(f"Training Features Shape: {X_train.shape}")
print(f"Testing Features Shape: {X_test.shape}")
print(f"Training Labels (Target) Shape: {y_train.shape}")
```
