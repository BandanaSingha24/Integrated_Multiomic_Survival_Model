```python
# Step 1: Data Cleaning and Indexing

import pandas as pd

# Re-read the CSV to ensure 'df' is in its initial state for processing
df = pd.read_csv('/content/processed_mrna_data.csv')

# The original df structure:
#        PATIENT_ID      RERE       RNF165        PHF7      CIDEA        TENT2
# 0  Entrez_Gene_Id  473.0000  494470.0000  51533.0000  1149.0000  167153.0000
# 1         MB-0362   -0.7139      -0.4606     -0.3325    -0.0129      -0.7853
# ...

# Extract the header row which contains gene IDs and the index name
header_row = df.iloc[0]

# The name for the index will be 'Entrez_Gene_Id'
index_col_name = header_row['PATIENT_ID'] # This is 'Entrez_Gene_Id'

# The actual gene IDs for column names are from the second element of the header row onwards
gene_ids = header_row.iloc[1:].astype(str).tolist()

# The data starts from the second row (index 1)
df_data = df.iloc[1:].copy()

# Rename the 'PATIENT_ID' column to the extracted index_col_name ('Entrez_Gene_Id')
# This column now holds the patient IDs (e.g., 'MB-0362')
df_data = df_data.rename(columns={'PATIENT_ID': index_col_name})

# Set the columns for the gene expression data
# The first column in df_data is now 'Entrez_Gene_Id' (containing patient IDs)
df_data.columns = [index_col_name] + gene_ids

# Set the 'Entrez_Gene_Id' column as the index
df = df_data.set_index(index_col_name)

# Convert all expression columns to numeric. Use errors='coerce' to turn non-numeric into NaN.
df = df.apply(pd.to_numeric, errors='coerce')

# Check if cleaning is successful
# देखें कि क्या सफाई सफल रही
print("Data shape after cleaning:", df.shape)
print(df.head())
```

```python
# Step 2: Log Transformation

import numpy as np

# Apply Log2 transformation to stabilize variance
df_log = np.log2(df + 1)

print("Log transformation completed successfully!")
print("First 5 rows of transformed data:")
print(df_log.head())
```

```python
# Step 2: Handling missing values and Log Transformation

import numpy as np

# 1. Fill missing values with 0 before calculation
df_filled = df.fillna(0)

# 2. Apply Log2 transformation to stabilize variance
df_log = np.log2(df_filled + 1)

# 3. Check for any remaining NaN values
print(f"Missing values remaining: {df_log.isnull().sum().sum()}")

# 4. Display the first 5 rows of the transformed data
print(df_log.head())
```

```python
# Step 3: Removing remaining NaN values

# 1. Option A: Replace all remaining NaNs with 0 (Standard for gene expression)
df_cleaned = df_log.fillna(0)

# 2. Check again if any NaNs are lef
remaining_nans = df_cleaned.isnull().sum().sum()
print(f"Remaining NaN values after filling: {remaining_nans}")

# 3. Update the main dataframe
df_final = df_cleaned

print("Data is now clean and ready for next steps!")
```

```python
# Step 4: Low Variance Filtering

# 1. Calculate the variance of each gene (column)
variance = df_final.var()

# 2. Set a threshold (e.g., keep genes with variance above 0.5)
threshold = 0.5
df_filtered = df_final.loc[:, variance > threshold]

# 3. Check the reduction in number of gene
print(f"Original gene count: {df_final.shape[1]}")
print(f"Filtered gene count: {df_filtered.shape[1]}")
print("Low variance filtering completed!")
```

```python
# Step 5: Standardization (Z-score scaling)

from sklearn.preprocessing import StandardScaler

# 1. Initialize the scaler
scaler = StandardScaler()

# 2. Fit and transform the filtered data
df_standardized = pd.DataFrame(scaler.fit_transform(df_filtered),
                               columns=df_filtered.columns,
                               index=df_filtered.index)

print("Standardization completed!")
print(df_standardized.head())
```

```python
# Save the processed matrix
df_standardized.to_csv('processed_gene_expression_feature_engineering.csv')
print("File saved successfully!")
```
