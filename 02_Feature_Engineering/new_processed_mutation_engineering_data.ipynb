```python
import pandas as pd
df=pd.read_csv("/content/md-1_processed_Transcriptomics_data.csv")
print (df.head())
```

```python
# step 1:RNA-seq cleaning and scaling

# 1. Identify and drop uncharacterised/pseudo-gene 'LOC' columns to reduce noise
loc_cols = [col for col in df.columns if col.startswith('LOC')]
df_filtered = df.drop(columns=loc_cols)
print(f"Dropped {len(loc_cols)} uncharacterized 'LOC' features to maximize biological signals.")

# 2. Separate patient keys and gene features
gene_features = [col for col in df_filtered.columns if col != 'PATIENT_ID']

# 3. Final Z-score scaling check to align scales exactly with clinical/mutation metrics
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_filtered[gene_features] = scaler.fit_transform(df_filtered[gene_features])

# 4. Save and verify your final engineered transcriptomics output
df_mrna_ready = df_filtered.copy()
print("Final Cleaned Transcriptomics Matrix Shape:", df_mrna_ready.shape)
```

```python
from google.colab import files

df_mrna_ready.to_csv('md-2_processed_RNA-seq_engineering_data.csv', index=False)

files.download('md-2_processed_RNA-seq_engineering_data.csv')
```
