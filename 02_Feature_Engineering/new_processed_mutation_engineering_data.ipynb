```python
import pandas as pd
df=pd.read_csv("/content/md-1_processed_mutation_data.csv")

print (df.head())
```

```python
# 1. Identify only the gene columns
gene_only_cols = [col for col in df.columns if col != 'PATIENT_ID']

# 2. Engineer the TMB feature as an additional column
df['TUMOR_MUTATIONAL_BURDEN'] = df[gene_only_cols].sum(axis=1)

# 3. Verify the new feature is added alongside individual genes
print("New shape with engineered feature:", df.shape)
print(df[['PATIENT_ID', 'TUMOR_MUTATIONAL_BURDEN']].head())
```
