```python
import pandas as Pd
df=pd.read_csv("/content/processed_mutation_matrix.csv")
print(df.head())
```

```python
# Step 1: Data Cleaning and Filtering

import pandas as pd

# Removing genes that have no mutations (all values are 0)
df_cleaned = df.loc[:, (df != 0).any(axis=0)]

# Checking for any missing values (NaN) in the dataset
missing_values = df_cleaned.isnull().sum().sum()

# Displaying the result
print(f"Total missing values: {missing_values}")
print("Shape of cleaned data:", df_cleaned.shape)
df_cleaned.head()
```

```python
# Step 2: Converting cleaned data into a Binary Matrix

# Converting values to 1 if mutation exists (>0) and 0 otherwise
binary_matrix = (df_cleaned > 0).astype(int)

# Display the first few rows of the binary matrix
print("Binary Matrix (first 5 rows):")
print(binary_matrix.head())

# Verify the shape of the binary matrix
print("Shape of binary matrix:", binary_matrix.shape)
```

```python
binary_matrix.to_csv("processed_mutation_feature_engineered_data.csv", index=True)
```
